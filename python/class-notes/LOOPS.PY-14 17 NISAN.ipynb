# WHILE LOOPS :
While loops have a boolean logic, similar to if statements. As long as the result of the condition returns True, the code block under while loop runs. When the condition returns to False, the loop execution is terminated and the program control moves further to the next operation.

````python
number = 0

while number < 6:
    print(number)
    number += 1
print('now, number is bigger or equal to 6')
````

````python
# Contine statement
i = 0
while i < 6:
  i += 1
  if i == 3:
    continue
  print(i)
````

````python
# Else statement : Print a message once the condition is false:
i = 1
while i < 6:
  print(i)
  i += 1
else:
  print("i is no longer less than 6")
````

````python
# Break ifadesiyle, tüm öğeler arasında döngü oluşturmadan önce döngüyü durdurabiliriz
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  print(x)
  if x == "banana":
    break
````

````python
# Continue deyimi ile döngünün mevcut yinelemesini durdurabilir ve bir sonrakine devam edebiliriz
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  if x == "banana":
    continue
  print(x)
````

````python
condition = True

if condition :
    a = 1
else :
    a = 0
    
print(a)
````

````python
# Yani body'ler en sagda ve solda yer alir.
a = 1 if 2 < 4 else 0
print(a)
print("---------")

print(1 if 2 < 4 else 0) # Degiskenleri ortadan kaldirarak yazdik.Expression en sonda bulunmaktadir gene ! 

print("--------")
print(1 if 5 < 4 else 0)
````
