     BASIC OPERATIONS WITH LISTS
     
     Listeler mutable veri tiplerdir. Yarinda degistirilebilirler.Dile dahil (built-in) fonksiyonlarƒ±n yanƒ± sƒ±ra, veri tiplerine √∂zg√º fonksiyonlar (‚Äúmetodlar‚Äù) da vardƒ±r. Metodlar sƒ±nƒ±f (‚Äúclass‚Äù) tanƒ±mƒ±ndaki i√ß deƒüi≈ükenlere eri≈üir, gerektiƒüinde deƒüi≈üiklik yaparlar. Metodlara eri≈üim ‚Äú.‚Äù (nokta) operat√∂r√ºyle saƒülanƒ±r; s√∂zgeli≈üi L bir listeyse, L.append(x) i≈ülemi L‚Äònin sonuna x‚Äòi ekler. Bu kƒ±sƒ±mda liste nesnelerinin yaygƒ±n olarak kullanƒ±lan metodlarƒ±ndan bahsedelim.


    Herhangi bir veri tipi i√ßinde hangi isimlerin (deƒüi≈üken veya metod) tanƒ±mlanmƒ±≈ü olduƒüunu dir fonksiyonuyla g√∂rebilirsiniz.
 
     .append() : Listenin sonunua eleman ekler.
     
     .insert() : Verilen bir konumdan bir √∂nceki konuma bir nesne ekler.
     
     .count()  : Listede belli bir deƒüere e≈üit ka√ß eleman olduƒüunu verir.
     
     .extend() : Listenin sonuna yeni bir liste ekleyerek geni≈ületir.append listenin sonuna tek eleman eklerken, extend yeni listenin t√ºm elemanlarƒ±nƒ± ekler.
     
     .index()  : Belli bir deƒüere sahip ilk elemanƒ±n bulunduƒüu konumu verir. Eleman listede yoksa ValueError hatasƒ± √ßƒ±karƒ±r.
     
     .pop()    : Belli bir konumdaki elemanƒ±n deƒüerini verir ve listeden siler. Arg√ºman almazsa, son elemanƒ± siler.
     
     .remove() : Belli bir deƒüere sahip ilk elemanƒ± listeden siler. Deƒüer listede yoksa ValueError hatasƒ± verir.Parantez icerisine silmek istedigimiz value'yi yazariz.
     
     .reverse(): Listeyi sondan ba≈üa ters √ßevirir. Dikkat: Yeni bir liste olu≈üturulmaz, orijinal liste deƒüi≈ütirilir. Eƒüer listedeki verileri bozmak istemiyorsanƒ±z √∂nce listeyi ba≈üka bir listeye kopyalamalƒ±sƒ±nƒ±z.
     
     .sort()   : isteyi sƒ±ralar ve orijinal listeyi deƒüi≈ütirir. Eƒüer listeyi bozmak istemiyorsanƒ±z, yukarƒ±daki gibi orijinal listenin bir kopyasƒ±nƒ± √ßƒ±karmanƒ±z gerekir.Parametreleri sorted fonksiyonuyla aynƒ±dƒ±r. Farkƒ±, sorted yeni bir liste verirken, sort metodunun orijinal listeyi deƒüi≈ütirmesidir. FARKLI TIPTE VERILERI SIRALAMAZ. ARKA PLANDA < > OPERATORLERINI CALISTIRIR.ASCII KODLARA GORE SIRALAMA YAPAR. True ve False phytonda 1 , 0 olarak algilandigi icin siralamasi gene ona gore olur. TERSTEN SIRALAMAK ICIN SORT ICINDE TANIMLI REVERSE DEGERINI TRUE OLARAK DEGISTIRMELIYIZ.
     
     .min() / .max() : Bir liste i√ßerisindeki minimum ve maksimum deƒüeri ister sayƒ±sal ister alfabetik olarak alabiliriz.
     
     .del()    : Herhangi bir indeks numarasƒ±ndaki elemanƒ± silebiliriz.Indeks numarasƒ± vermezsek listeyi olduƒüu gibi siler.
     
     .clear()  : Ayrƒ±ca del komutu ile listeyi sildiƒüimiz gibi clear() metodu ile de listeyi silebiliriz. Ancak arada ki fark del ile obje referansƒ±da silindiƒüinde listeye ula≈ümak istediƒüimizde NameError alƒ±rken clear() metodu ile hata almayƒ±z √ß√ºnk√º listenin referansƒ± bellekte olmaya devam eder ancak i√ßi bo≈ü olur.
     
     *.copy() :  Burada adres kopyalamasƒ± deƒüil de sadece i√ßlerindeki bilgiler kopyalanmƒ±≈ütƒ±r.Asil liste uzerinde yaptigimiz degisikler kopyalanan listeyi etkilemez veya tam tersi bir durum soz konusus olmaz.
     
     list()  : Yeni bir liste olusturmak icin veya icine kopyalamak istedigmiz listenin ismini yazarak kopyalama islemi yapabiliriz.
     
     s *= n : Updates s with its contents repeated n times

     len()  : Python listelerinde eleman sayƒ±sƒ±nƒ± len() metodu ile √∂ƒürenebiliriz. Aynƒ± metodu string i√ßinde kullanƒ±p karakter sayƒ±sƒ±nƒ± √∂ƒürenebiliriz.

    LISTELERE ATAMA YAPMA / INDEX ,SLICING / ACCESSING LIST
    
    1.) INDEX ISLEMI : Indeksleme islemi yaparak listenin elemanlari erisebilir ve sonrasinda degistirebiliriz. Bunun icin ulasmak istedigimiz elemanin indeks numaraini koseli parantez icerisine yazmamiz gerekmektedir.Negatif indeksleme yapilabilir,negetif indekslemede -1 'den ve sondan baslayarak saymaya baslariz. 
    
    my_list[INDEKS NUMARASI] : Verdigimiz indeks numarasindaki elemani bize dondurur.
    
    my_list[INDEKS NUMARASI] = VALUE : Verdigimiz indeks numarasina yeni atadigimiz value'yi yazdirir. Ilk degeri listeden kaldirir.
     
    
    
    2.) SLICING ISLEMI : A mechanism to select a range of items from sequence types like list, tuple, strings etc. is known as slicing. LISTED SLICING ISLEMI YAPTIGIMIZ ZAMAN GENE LISTE TIPINDE VERI DONDURUR.!!! SLICING'DA HER ZAMAN KENDI CINSINDEN DEGER DONDURUR LIST STR VB. GIBI..
    
    my_list[:]: returns the full copy of the sequence

    my_list[start:] : returns elements from start to the end element

    my_list[:stop] : returns element from the 1st element to stop-1

    my_list[::step] : returns each element with a given step ,Negatif steplemede slicing sureci sagdan sola dogru olur.

```python
city = ["ankara","istanbul", "Lozan"]   # INDEKSLEME YAPARAK INDEKS NUMARASI SIFIR OLAN DEGERI YENI ATADIGIMIZ DEGER ILE DEGISTIRIYORUZ.
city[0] = True 
print(city)
city[2]             # INDEKS NUMARASI 2 OLAN DEGERI BIZE DONDURUR.
```
