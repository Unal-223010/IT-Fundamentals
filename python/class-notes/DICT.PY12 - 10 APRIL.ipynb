```python
# Main Operations with Dictionaries
# Accessing Items : dict'lerde indekleme yapmak icin parantez icine key yazilir.

thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
x = thisdict["model"]
x
```

```python
# There is also a method called GET() that will give you the same result
y = thisdict.get("model")
y
```

```python
# The keys() method will return a list of all the keys in the dictionary.
v = thisdict.keys()
v
```

```python
# The values() method will return a list of all the values in the dictionary.
z = thisdict.values()
z
```

```python
car = {
"brand": "Ford",
"model": "Mustang",
"year": 1964
}

g = car.keys()

print(g) #before the change

car["color"] = "white"

print(g) #after the change
```

```python
car = {
"brand": "Ford",
"model": "Mustang",
"year": 1964
}

e = car.values()

print(e) #before the change

car["year"] = 2020

print(e) #after the change
```

```python
# Copy a Dictionary : You cannot copy a dictionary simply by typing dict2 = dict1, because: dict2 will only be a reference to dict1, and changes made in dict1 will automatically also be made in dict2.
#There are ways to make a copy, one way is to use the built-in Dictionary method copy().

thisdict = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
mydict = thisdict.copy()
print(mydict)
```
